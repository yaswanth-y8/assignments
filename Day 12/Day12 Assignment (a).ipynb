```python
path = "code/data/Predict-The-Data-Scientists-Salary-In-India_Train_Dataset.csv"

df = pd.read_csv(path)
df.head()
```

```python
df.drop("Unnamed: 0",axis=1,inplace=True)
df
```

```python
df['key_skills']=df['key_skills'].fillna(df['key_skills'].mode()[0])
```

```python
df["job_description"]=df["job_description"].fillna(df["job_description"].mode()[0])
df.isnull().sum()
```

```python
df["experience"].unique()
```

```python
df[['min_experience', 'max_experience']] = df['experience'].str.replace(' yrs', '').str.split('-', expand=True).astype(int)
df.drop("experience",axis=1,inplace=True)
df
```

```python
ln = LabelEncoder()
df['location'] = ln.fit_transform(df['location'])
df['salary'] = ln.fit_transform(df['salary'])
df.head()
```

```python
X = df.loc[:,['company_name_encoded','job_desig','job_type','min_experience','max_experience','location']]
y = df['salary']
X
```

```python
gen = X['job_desig'].value_counts()[X['job_desig'].value_counts() >= 5].index
X['job_desig'] = X['job_desig'].apply(lambda x: x if x in gen else 'Other')
X.head()
```

```python
en = OneHotEncoder(sparse_output=False, drop='first')
job_encoded = en.fit_transform(X[['job_desig', 'job_type']])
encoded_df = pd.DataFrame(job_encoded, columns=en.get_feature_names_out(['job_desig', 'job_type']))
```

```python
encoded_df.index = X.index
X = pd.concat([X.drop(columns=['job_desig', 'job_type']), encoded_df], axis=1)
```

```python
scaler = StandardScaler()
X[['min_experience', 'max_experience']] = scaler.fit_transform(X[['min_experience', 'max_experience']])
```

```python

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1,random_state = 42)
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
clf.score(x_train,y_train)
clf.score(x_test,y_test)
```

```python
ytestp = clf.predict(x_test)
ytrainp = clf.predict(x_train)
confusion_matrix(y_train,ytrainp)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, AdaBoostClassifier, GradientBoostingClassifier

def get_best_model(x_train, y_train, x_test, y_test):
    models = [
        DecisionTreeClassifier(),
        RandomForestClassifier(n_estimators=100),
        ExtraTreesClassifier(n_estimators=100),
        AdaBoostClassifier(n_estimators=100),
        GradientBoostingClassifier(n_estimators=100)
    ]
    best_model = None
    best_score = 0
    best_train = 0
    best_test = 0
    for model in models:
        model.fit(x_train, y_train)
        train_score = model.score(x_train, y_train)
        test_score = model.score(x_test, y_test)
        if test_score > best_score:
            best_score = test_score
            best_model = model
            best_train = train_score
            best_test = test_score
    return best_model, best_train, best_test
```

```python
best_estimator,train_acc,test_acc=get_best_model(x_train,y_train,x_test,y_test)
```
