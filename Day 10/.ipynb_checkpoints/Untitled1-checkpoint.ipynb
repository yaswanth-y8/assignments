```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("bikes.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.plot(df['Berri1'], label='Berri1', color='blue')
plt.plot(df['PierDup'], label='PierDup', color='red')
plt.legend()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df['Weekday'] = df.index.day_name()
print(df[['Berri1', 'PierDup', 'Weekday']].head())
```
