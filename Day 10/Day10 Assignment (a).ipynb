```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("code/data/bikes.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
plt.figure(figsize=(5, 4))
plt.plot(data['Berri1'], label='Berri1')
plt.plot(data['PierDup'], label='PierDup')

plt.title('Daily Attendance: Berri1 vs PierDup')
plt.xlabel('Date')
plt.ylabel('Attendance')
plt.legend()
plt.show()
```

```python
data['Date'] = pd.to_datetime(data['Date'],dayfirst=True)
```

```python
data.set_index('Date', inplace=True)
```

```python
data['weekday'] = data.index.day_name()
```

```python
print(data[['Berri1', 'PierDup','CSC','Mais1','Mais2','Parc','Rachel1','Totem_Laurier', 'weekday']])
```

```python
weekday_sum = data.groupby('weekday')[['Berri1', 'PierDup','CSC','Mais1','Mais2','Parc','Rachel1','Totem_Laurier']].sum()
```

```python
print(weekday_sum)
```

```python
weekday_sum.plot(kind='bar')
plt.title('Total Attendance by Weekday')
plt.xlabel('Weekday')
plt.ylabel('Total Attendance')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Track')
plt.tight_layout()
plt.show()
```
