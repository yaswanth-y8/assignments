```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x1 = np.random.normal(0, 1, 10000)
x2 = np.random.normal(0, 1, 10000)

x_sum = x1 + x2
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.hist(x1, bins=50, density=True, alpha=0.6, label='x1')
plt.title('x1 hist')

plt.subplot(1, 3, 2)
plt.hist(x2, bins=50, density=True, alpha=0.6, label='x2')
plt.title('x2 hist')

plt.subplot(1, 3, 3)
plt.hist(x_sum, bins=50, density=True, alpha=0.6, label='x1 + x2')
plt.title('x1 + x2  hist')
```
